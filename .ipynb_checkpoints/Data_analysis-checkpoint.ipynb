```python
# count nb of rows in view dataframe for each item id 
views_df = view.groupby('itemId').size()
clicks_df = click.groupby('itemId').size()
```

```python
ratios = []
n = 0
for item in clicks_df.keys():
    nbViews = views_df.get(item, 0)
    nbClicks = clicks_df.get(item, 0)
    ratio = nbViews / nbClicks 
    if ratio > 1:
        n += 1
    ratios.append(ratio)
print(n, len(clicks_df.keys()), n / len(clicks_df.keys()))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.hist(ratios, bins = np.array(list(range(0,20)) )/10, log=True)
plt.title('Distribution of ratio between number of clicks and views')
```

```python
# count nb of rows in view dataframe for each item id 
views_df = view.groupby('itemId').size()
purchases_df = purchases.groupby('itemId').size()
```

```python
ratios = []
for item in views_df.keys():
    nbViews = views_df.get(item, 0)
    nbPurchs = purchases_df.get(item, 0)
    ratio = nbPurchs / nbViews
    if ratio > 1:
        print(item, ': ', prod[prod['itemId'] == item]['pricelog2'])
    ratios.append(ratio)
```

```python
import numpy as np
```

```python
plt.hist(ratios, bins = np.array(list(range(0,10)) )/10, log=True)
plt.title('Distribution of ratio between number of views and purchases')
```
